```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_regression
```

```python
# n_samples means how many rows we want
# n_features=2 means num of column
# 
```

```python
x,y=make_regression(n_samples=10000,n_features=4,n_targets=1,noise=1000)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],y)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
```

```python
LR.fit(x,y)
DTR.fit(x,y)
```

```python
print(LR.score(x,y))
print(DTR.score(x,y))
```
